```python
import numpy as np
grp = df.groupby('prime_genre')
x = grp['user_rating'].agg(np.mean)
y = grp['price'].agg(np.sum)
z = grp['rating_count_tot'].agg(np.sum)

        
print(x)        
print(y)
print(z)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x,'ro', color='r')
plt.xticks(rotation=90)
plt.title('Category wise Rating')
plt.xlabel('prime_genre-->')
plt.ylabel('user_rating-->')
plt.show()
```

```python
cat_price = pd.cut(df['price'], pd.IntervalIndex.from_tuples([(-1.0,0.0),(0.1,5.0),(5.1,20.0),(20.1,500.0)]), 
                   labels=["Free","Low", "medium", "High"])
to_rep = ['(-1.0, 0.0]','(0.1, 5.0]','(5.1, 20.0]','(20.1, 500.0]']
rep_val = ["Free","Low ($0.1-$5.0)", "Medium ($5.1-$20.0)", "High ($20.1-500.0)"]

s=cat_price.apply(str)

df['cat_price'] = s.replace(to_replace=to_rep, value=rep_val)

df['cat_price'].value_counts().plot(kind='bar',figsize=(10,8),fontsize=18)
```

```python
df['cat_price'].value_counts()[0]/df['cat_price'].count()*100
```

```python
df['popularity'] = df['rating_count_tot'] * df['user_rating'] 

pop_free_games = df[df['prime_genre'] == 'Games'][df['cat_price']=='Free'][['track_name','popularity']].sort_values(by='popularity',ascending=False)[:10]


sns.barplot(x=pop_free_games['popularity'], y=pop_free_games['track_name'])
```
