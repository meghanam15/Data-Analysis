```python
df = df.dropna()
df
```

```python
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]),inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]),inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]),inplace=True)
```

```python

df.isnull().sum()
```

```python
df.info()
```

```python
df['Price'] = df['Price'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
df['Price'] = df['Price'].apply(lambda x: float(x))
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+', '') if '+' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',', '') if ',' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: float(x))
```

```python
grp = df.groupby('Category')
x = grp['Rating'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,5))
plt.plot(x, 'ro')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(x,'ro', color='r')
plt.xticks(rotation=90)
plt.title('Category wise Rating')
plt.xlabel('Categories-->')
plt.ylabel('Rating-->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(y,'r--', color='b')
plt.xticks(rotation=90)
plt.title('Category wise Pricing')
plt.xlabel('Categories-->')
plt.ylabel('Prices-->')
plt.show()
```

```python
cat_price = pd.cut(df['Price'], pd.IntervalIndex.from_tuples([(-1.0,0.0),(0.1,5.0),(5.1,20.0),(20.1,500.0)]), 
                   labels=["Free","Low", "medium", "High"])
to_rep = ['(-1.0, 0.0]','(0.1, 5.0]','(5.1, 20.0]','(20.1, 500.0]']
rep_val = ["Free","Low ($0.1-$5.0)", "Medium ($5.1-$20.0)", "High ($20.1-500.0)"]

s=cat_price.apply(str)

df['cat_price'] = s.replace(to_replace=to_rep, value=rep_val)

df['cat_price'].value_counts().plot(kind='bar',figsize=(10,8),fontsize=18)
```

```python
df['cat_price'].value_counts()[0]/df['cat_price'].count()*100
```

```python
df['popularity'] = df['Reviews'] * df['Rating'] 

pop_free_games= df[df['Category'] == 'FAMILY'][df['cat_price']=='Free'][['App','popularity']].sort_values(by='popularity',ascending=False)[:10]


sns.barplot(x=pop_free_games['popularity'], y=pop_free_games['App'])
```
